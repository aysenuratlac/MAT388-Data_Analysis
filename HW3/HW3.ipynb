First, let us load the data(there is no Nan values so no need imputation).Then define a function in order to do classification algorithms with ensemble learning strategies

```python
(mx_train, my_train), (mx_test, my_test)=tensorflow.keras.datasets.mnist.load_data()
```

```python
def classifier_model(X_train,X_test,y_train,y_test,strategy,model):
    res = strategy(model).fit(X_train, y_train)
    y_pred = res.predict(X_test)
    return classification_report(y_test, y_pred)
```

For this question use the World Bank's [Climate Change Knowledge Portal](https://climateknowledgeportal.worldbank.org/).

The Mediterranean Sea is bordered by the following 22 countries: Spain, France, Monaco, Italy, Slovenia, Croatia, Bosnia and Herzegovina, Montenegro, Albania, Greece, Turkey, Syria, Lebanon, Israel, Palestine, Egypt, Libya, Tunisia, Algeria, Morocco, Malta, and Cyprus. 

Using the World Bank's Climate Change Knowledge Portal, obtain the following data through their API:

- Monthly average precipitation

Gather this time series from 1950 for all countries surrounding the Mediterranean Sea. 

1. Write a function that takes 3 parameters:

   - 'surface' vs 'precipitation'

   that returns a

   - regular OLS regression

```python
countries = {
    'Spain': 'ESP',
    'France': 'FRA',
    'Monaco': 'MCO',
    'Italy': 'ITA',
    'Slovenia': 'SVN',
    'Croatia': 'HRV',
    'Bosnia and Herzegovina': 'BIH',
    'Montenegro': 'MNE',
    'Albania': 'ALB',
    'Greece': 'GRC',
    'Turkey': 'TUR',
    'Syria': 'SYR',
    'Lebanon': 'LBN',
    'Israel': 'ISR',
    'Palestine': 'PSE',
    'Egypt': 'EGY',
    'Libya': 'LBY',
    'Tunisia': 'TUN',
    'Algeria': 'DZA',
    'Morocco': 'MAR',
    'Malta': 'MLT',
    'Cyprus': 'CYP'
}
country_codes = ','.join(countries.values())
country_names = ','.join(countries.keys())
```

```python
api_url = f'https://cckpapi.worldbank.org/cckp/v1/era5-x0.5_timeseries_pr,tas_timeseries_monthly_1950-2020_mean_historical_era5_era5_mean/{country_codes}?_format=json'
```

```python
try:
    # Send a GET request to the API
    response = requests.get(api_url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # The response content will contain the data you requested
        data = response.json()  # or response.text if the response is in plain text
    else:
        # Print an error message for non-200 status codes
        print(f"Request failed with status code {response.status_code}")
except Exception as e:
    # Handle any exceptions that may occur during the request
    print(f"An error occurred: {str(e)}")
```

```python
df = pd.DataFrame(data)
```

Now, i defined a function below that takes data type, model type, and country name, then return the model which is constructed by using time and related data values. I thought seperating year and month might be better so i did. I also add the evaluation parts to the function(mae,mse, and r^2)

```python
def wb_getmodels(data_type,model_type,country_name):
    if data_type == 'surface':
        data = pd.DataFrame(df['data']['tas']).reset_index()
    elif data_type == 'precipitation':
        data = pd.DataFrame(df['data']['pr']).reset_index()
    else:
        print('invalid data parameter.')

    dates = pd.to_datetime(data['index'])
    year = dates.dt.year
    month = dates.dt.month
    data2=pd.DataFrame({'year': year, 'month': month,'Target': data[country_name]})
    X=data2[['year','month']]
    y = data2['Target']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

    # OLS
    if model_type == 'OLS':
        model = LinearRegression()
        model.fit(X_train, y_train)
    
    # Decision Tree Regression
    elif model_type == 'Decision Tree':
        model = DecisionTreeRegressor()
        model.fit(X_train, y_train)
    
    # SVM
    elif model_type == 'SVR':
        model = SVR()
        model.fit(X_train, y_train)
    
    # AdaBoost Regression
    elif model_type == 'AdaBoost':
        model = AdaBoostRegressor()
        model.fit(X_train, y_train)
    
    # LightGBM Regression
    elif model_type == 'LightGBM':
        model = LGBMRegressor()
        model.fit(X_train, y_train)
    
    # XGBoost Regression
    elif model_type == 'XGBoost':
        model = XGBRegressor()
        model.fit(X_train, y_train)
    else:
        print('Invalid model type')
    
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    return mae,mse,r2
```

Here are the lists of all countries and models:

```python
country_list=countries.values()
model_lists=['OLS','Decision Tree','SVR','AdaBoost','LightGBM','XGBoost']
```

There is a result dataframe which has the country, data type, model and scores. I iterate each data to each combination of country and model,then store the results in this dataframe:

```python
results_df = pd.DataFrame(columns=['Country','Data', 'Model', 'MAE', 'MSE', 'R2'])

# Iterate over countries and models
for country in country_list:
    for model in model_lists:
        # Call the function
        mae, mse, r2 = wb_getmodels('surface',model,country)
        # Append results to DataFrame
        results_df = results_df.append({'Country': country, 'Data': 'Avg surf. temp.','Model': model, 'MAE': mae, 'MSE': mse, 'R2': r2}, ignore_index=True)
```

```python
for country in country_list:
    for model in model_lists:
        # Call the function
        mae, mse, r2 = wb_getmodels('precipitation',model,country)
        # Append results to DataFrame
        results_df = results_df.append({'Country': country, 'Data': 'Avg prec.','Model': model, 'MAE': mae, 'MSE': mse, 'R2': r2}, ignore_index=True)
```

Now, we will display the results:

```python
# Display the results
grouped_results = results_df.groupby('Country')

# Iterate over groups and print results
for country, country_results in grouped_results:
    print(f"Results for {country}:")
    print(country_results)
    print("\n")
```
